5.    Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

> To generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap, you can follow these steps. We'll use NumPy to create the synthetic data and Seaborn to visualize the heatmap. Here's how you can do it:

Code Example

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(0)  # For reproducibility

# Create correlated features
size = 100  # Number of samples
x = np.random.normal(0, 1, size)
y = 2 * x + np.random.normal(0, 0.5, size)  # y is correlated with x
z = 3 * x + np.random.normal(0, 1, size)    # z is also correlated with x

# Create a DataFrame
data = pd.DataFrame({
    'Feature X': x,
    'Feature Y': y,
    'Feature Z': z
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, linewidths=0.5)

# Add titles and labels
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Explanation of the Code:

Synthetic Data Generation:

np.random.seed(0) ensures reproducibility.
x is generated from a normal distribution.
y and z are created as linear functions of x, making them correlated with x (with some added noise).

DataFrame Creation: The correlated features are stored in a Pandas DataFrame for easier manipulation and analysis.

Correlation Matrix Calculation: data.corr() computes the correlation matrix for the features in the DataFrame.

Heatmap Visualization:

sns.heatmap() creates a heatmap of the correlation matrix.
annot=True adds correlation coefficients to the cells.
cmap='coolwarm' specifies the color map for the heatmap.
fmt='.2f' formats the annotations to two decimal places.
Display: The heatmap is displayed with plt.show().

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sales Data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table for the heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt='.1f', cmap='YlGnBu', cbar_kws={'label': 'Average Sales'})

# Add titles and labels
plt.title('Heatmap of Average Sales Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Show the plot
plt.show()
```

Code Example

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output in a Jupyter notebook (remove if running as a script)
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = np.sin(x)  # Calculate the sine of x

# Create a Bokeh plot
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)', width=800, height=400) # Changed plot_width to width and plot_height to height

# Add a line renderer for the sine wave
p.line(x, y, legend_label='sin(x)', line_width=2, line_color='blue')

# Show the plot
show(p)
```

4.    Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

> To create a Bokeh histogram to visualize the distribution of the given data, you can use the following code. This code generates random data, computes the histogram, and then creates a Bokeh plot to display it.

Code Example

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output in a Jupyter notebook (remove if running as a script)
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh histogram
p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency',
           height=400, width=600) # Changed plot_height to height and plot_width to width

# Add bars to the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color='blue', line_color='white', alpha=0.7)

# Show the plot
show(p)
```

Explanation of the Code:

Imports:

The necessary libraries are imported: bokeh.plotting for plotting and numpy for numerical operations.

Output Setup:

output_notebook() is used to display the plot inline in a Jupyter notebook. You can remove this line if running the code in a standalone script.

Data Generation:

data_hist = np.random.randn(1000) generates 1000 random samples from a standard normal distribution.

Histogram Calculation:

np.histogram(data_hist, bins=30) computes the histogram of the data, returning the frequency counts (hist) and the bin edges (edges).

Bokeh Plot Creation:

A Bokeh figure is created with a title and axis labels.

Add Bars:

p.quad() is used to create the histogram bars. The top parameter specifies the heights of the bars, while left and right specify the edges of the bars.
The fill_color, line_color, and alpha parameters set the appearance of the bars.
Display the Plot:

show(p) renders the histogram.

> To create a Bokeh heatmap using the provided dataset, you can use the following code. This code generates random data, sets up the x and y coordinates, and creates a heatmap to visualize the data.

Code Example

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data_heatmap = np.random.rand(10, 10)

# Create x and y values
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(data_heatmap, extent=[0, 1, 0, 1], origin='lower', cmap='viridis')
plt.colorbar(label='Value')
plt.title('Heatmap Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
